```python
import json
import os
from tqdm import tqdm

def process_json(questions_path, images_dir, n=1000):
    # Cargar las preguntas de validación o entrenamiento
    with open(questions_path, "r") as f:
        questions = json.load(f)

    # Crear un diccionario para almacenar las imágenes disponibles en el directorio
    available_images = {img: os.path.join(images_dir, img) for img in os.listdir(images_dir)}
    print(f"Directorio de imágenes: {images_dir}")
    print(f"Total de imágenes disponibles: {len(available_images)}")

    # Crear una lista de ejemplos procesados
    data = {}
    data["images"] = []
    data["questions"] = []
    data["answers"] = []

    # Procesa cada pregunta y busca la imagen correspondiente
    for question in tqdm(questions["questions"][:n]):  # Procesa solo 'n' preguntas
        image_filename = question.get("image_filename")  # Extrae el nombre de la imagen
        if image_filename and image_filename in available_images:  # Verifica si la imagen está en el diccionario
            image_path = available_images[image_filename]
            data["images"].append(image_path)
            data["questions"].append(question["question"])
            data["answers"].append(question["answer"])
        else:
            print(f"Imagen no encontrada: {image_filename}")  # Mensaje de depuración

    print(f"Total de ejemplos procesados: {len(data['images'])}")
    return data
```

# Inference with BLIP (pre-trained, no fine-tuning)

```python
import torch
import requests
from PIL import Image
from transformers import BlipProcessor, BlipForQuestionAnswering

processor = BlipProcessor.from_pretrained("ybelkada/blip-vqa-base")
model = BlipForQuestionAnswering.from_pretrained("ybelkada/blip-vqa-base", torch_dtype=torch.float16).to("cuda")
```

```python
import matplotlib.pyplot as plt
from PIL import Image
import torch
import os

# Lista de índices para ejemplos de visualización
indices = [10, 11, 12, 13]  # Puedes ajustar estos valores para ver otras imágenes

# Definir la ruta base correcta para las imágenes
base_image_dir = "BD_entrenamiento/CLEVR_sample_10_000/images"

# Lista de carpetas donde se puede buscar la imagen
possible_subdirs = ["combined_val_images", "trainA", "valA", "valB"]

for i in indices:
    # Verificar que el índice no exceda el número de imágenes disponibles
    if i >= len(val_data['images']):
        print(f"Índice {i} fuera de rango. Saltando...")
        continue

    # Obtener la ruta de la imagen
    image_path = val_data['images'][i]

    # Verificar si la imagen existe en la ruta especificada, de no ser así, buscar en las subcarpetas
    if not os.path.exists(image_path):
        found = False
        image_filename = os.path.basename(image_path)

        for subdir in possible_subdirs:
            new_path = os.path.join(base_image_dir, subdir, image_filename)
            if os.path.exists(new_path):
                image_path = new_path
                found = True
                break

        if not found:
            print(f"No se encontró la imagen en ninguna ruta: {image_filename}")
            continue

    # Intentar cargar la imagen
    try:
        raw_image = Image.open(image_path).convert('RGB')
    except FileNotFoundError:
        print(f"No se encontró la imagen en la ruta: {image_path}")
        continue

    # Procesar la pregunta correspondiente
    question = val_data['questions'][i]
    inputs = processor(raw_image, question, return_tensors="pt").to("cuda", torch.float16)

    # Generar la respuesta del modelo
    out = model.generate(**inputs)

    # Mostrar la imagen y el resultado de inferencia
    plt.imshow(raw_image)
    plt.axis('off')
    plt.show()
    print('Question:', question)
    print('Model Answer:', processor.decode(out[0], skip_special_tokens=True))
    print('Correct Answer:', val_data['answers'][i])
```

```python
from collections import defaultdict
#default dict defaultimng to the same str used as key
synonyms = defaultdict(str)
synonyms['zero'] = '0'
synonyms['one'] = '1'
synonyms['two'] = '2'
synonyms['three'] = '3'
synonyms['four'] = '4'
synonyms['five'] = '5'
synonyms['rubber'] = 'plastic'
synonyms['rubber'] = 'plastic'
synonyms['yes'] #default --> ''
```

```python
def acc_metric(answers, gold):
    correct = 0
    for i in range(len(answers)):
        # Verificar si las respuestas coinciden o si un sinónimo coincide con la respuesta correcta
        if answers[i] == gold[i]:
            correct += 1
        elif synonyms[answers[i]] == gold[i]:  # Usa el sinónimo si existe
            correct += 1
    return correct / len(answers)
```

```python
model_ans = []
base_image_dir = "BD_entrenamiento/CLEVR_sample_10_000/images"
possible_subdirs = ["combined_val_images", "trainA", "valA", "valB"]

for i in range(len(val_data['images'])):
    # Obtener la ruta de la imagen y verificar si existe
    image_path = val_data['images'][i]

    if not os.path.exists(image_path):
        found = False
        image_filename = os.path.basename(image_path)

        # Intentar buscar la imagen en las subcarpetas
        for subdir in possible_subdirs:
            new_path = os.path.join(base_image_dir, subdir, image_filename)
            if os.path.exists(new_path):
                image_path = new_path
                found = True
                break

        if not found:
            print(f"No se encontró la imagen en ninguna ruta: {image_filename}")
            model_ans.append("Imagen no encontrada")
            continue

    # Cargar la imagen y realizar la inferencia
    raw_image = Image.open(image_path).convert('RGB')
    question = val_data['questions'][i]
    inputs = processor(raw_image, question, return_tensors="pt").to("cuda", torch.float16)
    out = model.generate(**inputs)

    # Decodificar la respuesta y agregarla a model_ans
    model_ans.append(processor.decode(out[0], skip_special_tokens=True))

print("Respuestas generadas para todas las imágenes.")
```

```python
acc = acc_metric(model_ans, val_data['answers'])
print(f'Accuracy: {acc*100}%')
```

```python
def conventional_accuracy(predictions, ground_truth):
    correct = 0
    total = len(predictions)

    for i in range(total):
        if predictions[i] == ground_truth[i]:
            correct += 1

    accuracy = correct / total
    return accuracy

# Calcular la precisión de manera convencional
acc_conventional = conventional_accuracy(model_ans, val_data['answers'])
print(f'Conventional Accuracy: {acc_conventional * 100:.2f}%')
```

```python
from datasets import Dataset

train_dataset = Dataset.from_dict(train_data)
val_dataset = Dataset.from_dict(val_data)
#test_dataset = Dataset.from_dict(test_data)
```

```python
def preprocess_data(examples):
    base_image_dir = "BD_entrenamiento/CLEVR_sample_10_000/images"
    possible_subdirs = ["combined_val_images", "trainA", "valA", "valB"]

    # List to store valid image paths
    image_paths = []

    # Buscar y cargar cada imagen
    for image_path in examples["images"]:
        # Si la ruta inicial no existe, buscar en las subcarpetas
        if not os.path.exists(image_path):
            found = False
            image_filename = os.path.basename(image_path)

            for subdir in possible_subdirs:
                new_path = os.path.join(base_image_dir, subdir, image_filename)
                if os.path.exists(new_path):
                    image_path = new_path
                    found = True
                    break

            if not found:
                print(f"No se encontró la imagen en ninguna ruta: {image_filename}")
                image_paths.append(None)  # Placeholder para manejar el error
                continue

        # Añadir la ruta de la imagen válida a la lista
        image_paths.append(image_path)

    # Filtrar imágenes no encontradas
    valid_images = [Image.open(path).convert("RGB") for path in image_paths if path is not None]
    questions = [q for i, q in enumerate(examples["questions"]) if image_paths[i] is not None]
    answers = [a for i, a in enumerate(examples["answers"]) if image_paths[i] is not None]

    # Tokenizar las entradas
    inputs = processor(images=valid_images, text=questions, return_tensors="pt", padding=True)

    # Añadir etiquetas (respuestas)
    inputs['labels'] = processor.tokenizer(answers, padding=True, truncation=True, return_tensors="pt").input_ids
    return inputs
```

```python
processed_train_dataset = train_dataset.map(preprocess_data, batched=True)
processed_val_dataset = val_dataset.map(preprocess_data, batched=True)
# processed_test_dataset = test_dataset.map(preprocess_data, batched=True)  # Descomenta si tienes test_data
```
